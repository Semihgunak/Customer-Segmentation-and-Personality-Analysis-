```python
import pandas as pd

def convert_to_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%d/%m/%Y')
    except ValueError:
        return pd.to_datetime(date_str, format='%d-%m-%Y')

df["Dt_Customer"] = df["Dt_Customer"].apply(convert_to_date)

dates = df["Dt_Customer"]

print("The newest customer's enrolment date in the records:", max(dates))
print("The oldest customer's enrolment date in the records:", min(dates))
```

```python
days = []
d1 = max(dates)
for i in dates:
    delta = d1 - i
    days.append(delta)
df["Customer_For"] = days
df['Customer_For'] = df['Customer_For'].apply(lambda x:x.days)
```

```python
df = df[df['Customer_Age'] < 90]
df = df[df['Income'] < 300000]
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df['Education'] = label_encoder.fit_transform(df['Education'])
df['Marital_Status'] = label_encoder.fit_transform(df['Marital_Status'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
col_scale = ['Income', 'Kids', 'Expenses',
       'TotalAcceptedCmp', 'NumTotalPurchases', 'Customer_Age', 'Customer_For']

df[col_scale] = scaler.fit_transform(df[col_scale])
```

```python
veri = df.copy()
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(veri)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(16,8))
plt.plot(range(1,11),wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=2, random_state=42).fit(veri)
pred=kmeans.predict(veri)
veri['cluster_Kmeans'] = pred + 1
```

```python
veri.head()
```

```python
sns.countplot(x=veri["cluster_Kmeans"])
plt.title("Distribution Of The Clusters")
plt.show()
```

```python
veri1= df.copy()
```

```python
from sklearn.decomposition import PCA
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(veri1)
PCA_ds = pd.DataFrame(pca.transform(veri1), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA

wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(PCA_ds)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(16,8))
plt.plot(range(1,11),wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
AC = AgglomerativeClustering(n_clusters=2)

yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = yhat_AC
veri1["Cluster_Agglo"]= yhat_AC + 1
```

```python
sns.countplot(x=veri1["Cluster_Agglo"])
plt.title("Distribution Of The Clusters")
plt.show()
```

```python
fig = plt.figure(figsize=(16,14))
ax = plt.subplot(111, projection='3d', label="bla")

ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o')
ax.set_title("The Plot Of The Clusters")
```

```python
pl = sns.scatterplot(data = data,y=veri1["Expenses"], x=veri1["Income"],hue=veri1["Cluster_Agglo"])
pl.set_title("Cluster's Profile Based On Income And Spending")
plt.legend()
plt.show()
```
